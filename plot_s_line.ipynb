```python
# -*- coding: utf-8 -*-
"""
Created on Thu Mar  4 10:11:05 2021

@author: John Meluso
"""
import matplotlib as mpl
import matplotlib.pylab as pylab


def set_fonts(extra_params={}):
    params = {
        "font.family": "Serif",
        # "font.sans-serif": ["Tahoma", "DejaVu Sans", "Lucida Grande", "Verdana"],
        "mathtext.fontset": "cm",
        "legend.fontsize": 14,
        "axes.labelsize": 14,
        "axes.titlesize": 14,
        "xtick.labelsize": 14,
        "ytick.labelsize": 14,
        "figure.titlesize": 14,
    }
    for key, value in extra_params.items():
        params[key] = value
    pylab.rcParams.update(params)
set_fonts()
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

colors = ["#2d5d83", "#4f9bd9", "#077187", "#3dfaff", "#8b888e"]

df = pd.read_csv("results/s_count.txt", sep="\t")

plt.figure(figsize=(11, 5), dpi=100)

plt.subplot(121)
ax = plt.gca()
plt.text(0.15, 0.15, "(a)", fontsize=20, transform=ax.transAxes)

plt.plot(df["s"], df["edges"], "o", color=colors[0], markersize=5)
plt.xlabel(r"$s$")
plt.ylabel("Number of Edges")
plt.yscale("log")
plt.xscale("log")
# plt.xticks(ticks=range(df["s"].min(), df["s"].max() + 1, 50))
sns.despine()

plt.subplot(122)
ax = plt.gca()
plt.text(0.15, 0.15, "(b)", fontsize=20, transform=ax.transAxes)

plt.plot(df["s"], df["nodes"], "o", color=colors[1], markersize=5)
plt.xlabel(r"$s$")
plt.ylabel("Number of Nodes")
plt.yscale("log")
plt.xscale("log")
# plt.xticks(ticks=range(df["s"].min(), df["s"].max() + 1, 50))
sns.despine()

plt.tight_layout()
plt.savefig("results/sline_graph_stats.pdf", dpi=1000)
```
